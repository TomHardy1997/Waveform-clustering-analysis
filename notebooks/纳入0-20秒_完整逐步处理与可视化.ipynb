- 刺激期：5-10 秒

## 2. 设置路径和可修改参数

后续最常改的是这里。  
如果实验老师想让峰更“严格”，可以提高 `PEAK_HEIGHT` 或 `PEAK_PROMINENCE`。  
如果想让峰更“敏感”，可以降低这两个值。

```python
PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
ROOT = PROJECT_ROOT
VIS_DIR = PROJECT_ROOT / 'results' / 'visualization'
LONG_CSV = VIS_DIR / '纳入文件_0-20秒原始波形长表.csv'

RESULT_DIR = PROJECT_ROOT / 'results' / 'preprocessing'
NOTEBOOK_OUT_DIR = RESULT_DIR / 'notebook逐步运行结果'
CSV_DIR = NOTEBOOK_OUT_DIR / '单文件CSV数据'
PLOT_DIR = NOTEBOOK_OUT_DIR / '单文件预处理波形图'
OVERVIEW_DIR = NOTEBOOK_OUT_DIR / '总览图'

for folder in [NOTEBOOK_OUT_DIR, CSV_DIR, PLOT_DIR, OVERVIEW_DIR]:
    folder.mkdir(parents=True, exist_ok=True)

# 时间窗口。现在只分析 0-20 秒。
STIM_ON = 5.0
STIM_OFF = 10.0
COMMON_END = 20.0

# 预处理参数。
BASELINE_RATIO = 0.25       # 用前 25% 的点估计基线，也就是约 0-5 秒
SAVGOL_WINDOW = 51          # 平滑窗口，必须是奇数；越大越平滑
SAVGOL_POLYORDER = 3        # 平滑多项式阶数

# 波峰判断参数。
PEAK_HEIGHT = 0.003         # 峰顶至少高于基线 0.003 ΔF/F0
PEAK_PROMINENCE = 0.002     # 峰相对左右背景至少突出 0.002
PEAK_DISTANCE_SECONDS = 0.30 # 两个峰之间至少间隔 0.30 秒

# 波谷判断参数。
VALLEY_DEPTH = 0.003        # 谷底至少低于基线 -0.003 ΔF/F0
VALLEY_PROMINENCE = 0.002   # 波谷相对左右背景至少突出 0.002

print('输入长表:', LONG_CSV)
print('本 notebook 输出:', NOTEBOOK_OUT_DIR)
```

## 3. 波峰和波谷的判断标准

本轮不是直接在原始曲线上找峰，而是在预处理后的 `基线校正后 ΔF/F0` 上判断。

预处理顺序：

1. 对原始 `F/F0` 做 3 倍标准差截断，减少极端异常点影响。
2. 用前 25% 时间点作为基线，计算基线值。
3. 将曲线除以基线，得到基线校正后的 `F/F0`。
4. 用 Savitzky-Golay 方法平滑。
5. 再减去 1，得到围绕 0 上下波动的 `ΔF/F0`。

波峰需要同时满足：

- 峰值高度 `height >= 0.003`。
- 峰突出度 `prominence >= 0.002`。
- 相邻两个波峰至少间隔 `0.30 秒`。

波谷的判断方式是把曲线取负后再找峰，所以波谷需要满足：

- 谷底深度 `depth >= 0.003`，也就是原曲线中 `ΔF/F0 <= -0.003`。
- 谷突出度 `prominence >= 0.002`。
- 相邻两个波谷至少间隔 `0.30 秒`。

这几个阈值决定了“什么算一个明显峰/谷”。后续如果觉得峰太多，就提高阈值；如果漏掉了肉眼可见的小峰，就降低阈值。

## 5. 定义基础函数

```python
def ensure_odd_window(window, n):
    """保证平滑窗口是奇数，并且不超过当前曲线长度。"""
    window = int(window)
    if window % 2 == 0:
        window += 1
    if window >= n:
        window = n - 1 if n % 2 == 0 else n
    if window < 5:
        return None
    return window


def segment_mask(time, start, end):
    return (time >= start) & (time < end)


def safe_auc(time, y):
    if len(time) < 2:
        return np.nan
    return float(np.trapezoid(y, time))


def safe_slope(time, y):
    if len(time) < 2 or np.nanstd(time) == 0:
        return np.nan
    return float(np.polyfit(time - time[0], y, 1)[0])


def count_events(times, amps, start, end, strongest='max'):
    times = np.asarray(times)
    amps = np.asarray(amps)
    mask = (times >= start) & (times < end)
    t = times[mask]
    a = amps[mask]
    if len(a) == 0:
        return 0, 0.0, np.nan, np.nan
    idx = int(np.argmin(a) if strongest == 'min' else np.argmax(a))
    return int(len(a)), float(a[idx]), float(t[idx]), float(t[0])


def classify_response(row):
    on = row.get('ON反应幅度', 0)
    off = row.get('OFF反应幅度', 0)
    stim_delta = row.get('刺激期相对刺激前变化量', 0)
    if on > PEAK_HEIGHT and off > PEAK_HEIGHT:
        return 'ON-OFF型'
    if on > PEAK_HEIGHT:
        return 'ON型'
    if off > PEAK_HEIGHT:
        return 'OFF型'
    if stim_delta < -PEAK_HEIGHT:
        return '抑制型'
    return '弱反应或无反应型'
```

## 6. 单个文件的预处理、峰谷检测和特征提取

这一格是核心逻辑。输出包括：

- 预处理后的逐时间点数据。
- 每个波峰的时间、幅度、半高宽、上升/下降时间。
- 每个波谷的时间、幅度、半高宽。
- 分阶段特征和核心聚类特征。

```python
def segment_stats(prefix, time, y):
    ranges = {
        'pre_0_5s': (0, STIM_ON),
        'stim_5_10s': (STIM_ON, STIM_OFF),
        'post_10_20s': (STIM_OFF, COMMON_END + 1e-9),
    }
    start, end = ranges[prefix]
    mask = segment_mask(time, start, end)
    vals = y[mask]
    t = time[mask]
    if len(vals) == 0:
        return {
            f'{prefix}_mean': np.nan,
            f'{prefix}_max': np.nan,
            f'{prefix}_min': np.nan,
            f'{prefix}_std': np.nan,
            f'{prefix}_auc': np.nan,
            f'{prefix}_slope': np.nan,
        }
    return {
        f'{prefix}_mean': float(np.nanmean(vals)),
        f'{prefix}_max': float(np.nanmax(vals)),
        f'{prefix}_min': float(np.nanmin(vals)),
        f'{prefix}_std': float(np.nanstd(vals)),
        f'{prefix}_auc': safe_auc(t, vals),
        f'{prefix}_slope': safe_slope(t, vals),
    }


def preprocess_one(group):
    group = group.sort_values('time')
    file_id = group['file_id'].iloc[0]
    filename = group['filename'].iloc[0]

    time = group['time'].to_numpy(dtype=float)
    f_raw = group['f_f0'].to_numpy(dtype=float)
    dff_raw = group['dff'].to_numpy(dtype=float)
    if np.isnan(dff_raw).all():
        dff_raw = f_raw - 1

    dt = float(np.nanmedian(np.diff(time)))
    sample_rate = 1 / dt

    # 1. 异常值截断。
    mean_f = float(np.nanmean(f_raw))
    std_f = float(np.nanstd(f_raw))
    f_clip = np.clip(f_raw, mean_f - 3 * std_f, mean_f + 3 * std_f)

    # 2. 基线校正。
    baseline_length = max(3, int(len(f_clip) * BASELINE_RATIO))
    baseline_value = float(np.nanmean(f_clip[:baseline_length]))
    f_baseline = f_clip / baseline_value

    # 3. 平滑。
    window = ensure_odd_window(SAVGOL_WINDOW, len(f_baseline))
    if window is None:
        f_smooth = f_baseline.copy()
        dff_smooth = dff_raw.copy()
    else:
        f_smooth = savgol_filter(f_baseline, window_length=window, polyorder=SAVGOL_POLYORDER)
        dff_smooth = savgol_filter(dff_raw, window_length=window, polyorder=SAVGOL_POLYORDER)

    baseline_corrected = f_smooth - 1
    peak_distance = max(1, int(sample_rate * PEAK_DISTANCE_SECONDS))

    # 4. 波峰检测。
    peaks, _ = find_peaks(
        baseline_corrected,
        height=PEAK_HEIGHT,
        prominence=PEAK_PROMINENCE,
        distance=peak_distance,
    )

    # 5. 波谷检测：把曲线取负，然后找峰。
    valleys, _ = find_peaks(
        -baseline_corrected,
        height=VALLEY_DEPTH,
        prominence=VALLEY_PROMINENCE,
        distance=peak_distance,
    )

    preprocess = pd.DataFrame({
        'file_id': file_id,
        '文件名': filename,
        '时间(秒)': time,
        '原始F/F0': f_raw,
        '异常值截断后F/F0': f_clip,
        '基线校正后F/F0': f_baseline,
        '平滑去噪后F/F0': f_smooth,
        '基线校正后ΔF/F0': baseline_corrected,
        '原始dF/F0': dff_raw,
        '平滑去噪后dF/F0': dff_smooth,
    })

    peak_rows = []
    if len(peaks):
        widths, _, left_ips, right_ips = peak_widths(baseline_corrected, peaks, rel_height=0.5)
        for i, peak_idx in enumerate(peaks):
            amp = float(baseline_corrected[peak_idx])
            rise_threshold = 0.1 * amp
            left_candidates = np.where(baseline_corrected[:peak_idx] <= rise_threshold)[0]
            rise_start = int(left_candidates[-1]) if len(left_candidates) else 0
            right_candidates = np.where(baseline_corrected[peak_idx:] <= rise_threshold)[0]
            fall_end = int(right_candidates[0] + peak_idx) if len(right_candidates) else len(baseline_corrected) - 1
            rise_time = float(time[peak_idx] - time[rise_start])
            fall_time = float(time[fall_end] - time[peak_idx])
            peak_rows.append({
                'file_id': file_id,
                '文件名': filename,
                '峰编号': i + 1,
                '峰值时间(秒)': round(float(time[peak_idx]), 4),
                '峰值幅度(ΔF/F0)': round(amp, 6),
                '半高宽(秒)': round(float(widths[i] / sample_rate), 4),
                '半高宽左边界(秒)': round(float(time[0] + left_ips[i] / sample_rate), 4),
                '半高宽右边界(秒)': round(float(time[0] + right_ips[i] / sample_rate), 4),
                '10%-90%上升时间(秒)': round(rise_time, 4),
                '平均上升速率(ΔF/F0/秒)': round(float(amp / rise_time), 6) if rise_time > 0 else np.nan,
                '90%-10%下降时间(秒)': round(fall_time, 4),
                '平均下降速率(ΔF/F0/秒)': round(float(amp / fall_time), 6) if fall_time > 0 else np.nan,
            })
    peak_df = pd.DataFrame(peak_rows)

    valley_rows = []
    if len(valleys):
        widths, _, left_ips, right_ips = peak_widths(-baseline_corrected, valleys, rel_height=0.5)
        for i, valley_idx in enumerate(valleys):
            amp = float(baseline_corrected[valley_idx])
            valley_rows.append({
                'file_id': file_id,
                '文件名': filename,
                '谷编号': i + 1,
                '波谷时间(秒)': round(float(time[valley_idx]), 4),
                '波谷幅度(ΔF/F0)': round(amp, 6),
                '波谷半高宽(秒)': round(float(widths[i] / sample_rate), 4),
                '半高宽左边界(秒)': round(float(time[0] + left_ips[i] / sample_rate), 4),
                '半高宽右边界(秒)': round(float(time[0] + right_ips[i] / sample_rate), 4),
            })
    valley_df = pd.DataFrame(valley_rows)

    peak_times = np.array([r['峰值时间(秒)'] for r in peak_rows], dtype=float) if peak_rows else np.array([])
    peak_amps = np.array([r['峰值幅度(ΔF/F0)'] for r in peak_rows], dtype=float) if peak_rows else np.array([])
    valley_times = np.array([r['波谷时间(秒)'] for r in valley_rows], dtype=float) if valley_rows else np.array([])
    valley_amps = np.array([r['波谷幅度(ΔF/F0)'] for r in valley_rows], dtype=float) if valley_rows else np.array([])

    stage = {'file_id': file_id, '文件名': filename}
    for prefix in ['pre_0_5s', 'stim_5_10s', 'post_10_20s']:
        stage.update(segment_stats(prefix, time, baseline_corrected))

    for label, start, end in [
        ('pre_0_5s', 0, STIM_ON),
        ('stim_5_10s', STIM_ON, STIM_OFF),
        ('post_10_20s', STIM_OFF, COMMON_END + 1e-9),
    ]:
        pc, pa, pt, pfirst = count_events(peak_times, peak_amps, start, end, strongest='max')
        vc, va, vt, vfirst = count_events(valley_times, valley_amps, start, end, strongest='min')
        stage[f'{label}_peak_count'] = pc
        stage[f'{label}_max_peak_amp'] = pa
        stage[f'{label}_max_peak_time'] = pt
        stage[f'{label}_first_peak_time'] = pfirst
        stage[f'{label}_valley_count'] = vc
        stage[f'{label}_min_valley_amp'] = va
        stage[f'{label}_min_valley_time'] = vt
        stage[f'{label}_first_valley_time'] = vfirst

    pre_mean = stage['pre_0_5s_mean']
    stim_mean = stage['stim_5_10s_mean']
    post_mean = stage['post_10_20s_mean']

    on_count, on_amp, on_time, on_first = count_events(peak_times, peak_amps, STIM_ON, STIM_OFF, strongest='max')
    off_count, off_amp, off_time, off_first = count_events(peak_times, peak_amps, STIM_OFF, COMMON_END + 1e-9, strongest='max')

    intervals = np.diff(peak_times) if len(peak_times) >= 2 else np.array([])
    response_range = float(np.nanmax(baseline_corrected) - np.nanmin(baseline_corrected))
    response_std = float(np.nanstd(baseline_corrected))
    response_mean_abs = float(np.nanmean(np.abs(baseline_corrected)))

    summary = {
        'file_id': file_id,
        '文件名': filename,
        '总波峰数量': int(len(peak_times)),
        '总波谷数量': int(len(valley_times)),
        '最大波峰幅度': float(np.max(peak_amps)) if len(peak_amps) else 0.0,
        '最大波峰出现时间': float(peak_times[np.argmax(peak_amps)]) if len(peak_amps) else np.nan,
        '最大波谷幅度': float(np.min(valley_amps)) if len(valley_amps) else 0.0,
        '最大波谷出现时间': float(valley_times[np.argmin(valley_amps)]) if len(valley_amps) else np.nan,
        '平均波峰幅度': float(np.mean(peak_amps)) if len(peak_amps) else 0.0,
        '平均波谷幅度': float(np.mean(valley_amps)) if len(valley_amps) else 0.0,
        '波峰平均宽度': float(peak_df['半高宽(秒)'].mean()) if len(peak_df) else np.nan,
        '波峰平均上升时间': float(peak_df['10%-90%上升时间(秒)'].mean()) if len(peak_df) else np.nan,
        '波峰平均下降时间': float(peak_df['90%-10%下降时间(秒)'].mean()) if len(peak_df) else np.nan,
        '波峰发放频率': float(len(peak_times) / COMMON_END),
        '刺激前是否有波峰': int(stage['pre_0_5s_peak_count'] > 0),
        '刺激前是否有波谷': int(stage['pre_0_5s_valley_count'] > 0),
        '撤光后是否有波峰': int(stage['post_10_20s_peak_count'] > 0),
        '撤光后是否有波谷': int(stage['post_10_20s_valley_count'] > 0),
        'ON反应幅度': on_amp,
        'ON反应潜伏期': float(on_first - STIM_ON) if not np.isnan(on_first) else np.nan,
        'OFF反应幅度': off_amp,
        'OFF反应潜伏期': float(off_first - STIM_OFF) if not np.isnan(off_first) else np.nan,
        'ON/OFF比值': float(on_amp / off_amp) if off_amp > 0 else np.nan,
        '刺激前平均值': pre_mean,
        '刺激期平均值': stim_mean,
        '撤光后平均值': post_mean,
        '刺激期相对刺激前变化量': float(stim_mean - pre_mean),
        '撤光后相对刺激前变化量': float(post_mean - pre_mean),
        '曲线面积AUC': safe_auc(time, baseline_corrected),
        '刺激期AUC': stage['stim_5_10s_auc'],
        '撤光后AUC': stage['post_10_20s_auc'],
        '最大值和最小值差值': response_range,
        '波形标准差': response_std,
        '波形变异系数': float(response_std / response_mean_abs) if response_mean_abs > 0 else np.nan,
        '整体斜率': safe_slope(time, baseline_corrected),
        '恢复斜率': stage['post_10_20s_slope'],
        '第一个波峰出现时间': float(peak_times[0]) if len(peak_times) else np.nan,
        '最后一个波峰出现时间': float(peak_times[-1]) if len(peak_times) else np.nan,
        '第一个波谷出现时间': float(valley_times[0]) if len(valley_times) else np.nan,
        '峰间隔平均值': float(np.mean(intervals)) if len(intervals) else np.nan,
        '峰间隔变异性': float(np.std(intervals)) if len(intervals) else np.nan,
    }
    summary.update(stage)
    summary['反应类型初判'] = classify_response(summary)

    return preprocess, peak_df, valley_df, pd.DataFrame([stage]), pd.DataFrame([summary])
```

## 7. 批量处理全部细胞

这一步会遍历所有纳入文件。运行结束后会得到 5 张总表。

```python
preprocess_all = []
peaks_all = []
valleys_all = []
stages_all = []
summary_all = []

for i, (_, group) in enumerate(raw_long.groupby('file_id', sort=True), start=1):
    preprocess, peaks, valleys, stages, summary = preprocess_one(group)
    preprocess_all.append(preprocess)
    peaks_all.append(peaks)
    valleys_all.append(valleys)
    stages_all.append(stages)
    summary_all.append(summary)
    if i % 25 == 0:
        print(f'已处理 {i} 个文件')

preprocess_df = pd.concat(preprocess_all, ignore_index=True)
peaks_df = pd.concat(peaks_all, ignore_index=True) if peaks_all else pd.DataFrame()
valleys_df = pd.concat(valleys_all, ignore_index=True) if valleys_all else pd.DataFrame()
stages_df = pd.concat(stages_all, ignore_index=True)
summary_df = pd.concat(summary_all, ignore_index=True)

print('处理完成:', summary_df.shape)
display(summary_df.head())
```

## 11. 单个细胞画图函数

图里：

- 黑色线：基线校正后的 `ΔF/F0`。
- 红点：检测到的波峰。
- 蓝点：检测到的波谷。
- 黄色区域：5-10 秒刺激期。

```python
def plot_one_cell(file_id):
    one = preprocess_df[preprocess_df['file_id'] == file_id]
    one_peaks = peaks_df[peaks_df['file_id'] == file_id] if len(peaks_df) else pd.DataFrame()
    one_valleys = valleys_df[valleys_df['file_id'] == file_id] if len(valleys_df) else pd.DataFrame()

    fig, ax = plt.subplots(figsize=(10, 4))
    ax.plot(one['时间(秒)'], one['基线校正后ΔF/F0'], color='black', linewidth=1.5)
    ax.axhline(0, color='gray', linewidth=1)
    ax.axvspan(STIM_ON, STIM_OFF, color='#f2cc68', alpha=0.30, label='Stim 5-10s')

    if len(one_peaks):
        ax.scatter(one_peaks['峰值时间(秒)'], one_peaks['峰值幅度(ΔF/F0)'], color='red', s=36, label='Peaks')
    if len(one_valleys):
        ax.scatter(one_valleys['波谷时间(秒)'], one_valleys['波谷幅度(ΔF/F0)'], color='royalblue', s=36, label='Valleys')

    ax.set_title(str(file_id))
    ax.set_xlabel('Time (s)')
    ax.set_ylabel('baseline corrected ΔF/F0')
    ax.legend(loc='best')
    fig.tight_layout()
    return fig

example = summary_df.iloc[0]['file_id']
fig = plot_one_cell(example)
plt.show()
```

## 12. 批量保存每个细胞的波形图

```python
for i, file_id in enumerate(summary_df['file_id'], start=1):
    fig = plot_one_cell(file_id)
    fig.savefig(PLOT_DIR / f'{file_id}_预处理峰谷图.png', dpi=180)
    plt.close(fig)
    if i % 25 == 0:
        print(f'已保存 {i} 张图')

print('单细胞图保存到:', PLOT_DIR)
```
