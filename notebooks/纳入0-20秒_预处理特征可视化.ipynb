# 纳入 0-20 秒 Excel 的预处理、特征提取和可视化

这份 notebook 用来查看已经生成的结果。处理流程参考 `Processing_waveform.py` 和示例处理结果 Excel，并加入 0-5、5-10、10-20 秒阶段特征。


```python
from pathlib import Path
import pandas as pd
from IPython.display import Image, display

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
ROOT = PROJECT_ROOT
RESULT_DIR = PROJECT_ROOT / 'results' / 'preprocessing'
CSV_DIR = RESULT_DIR / '单文件CSV数据'
PLOT_DIR = RESULT_DIR / '单文件预处理波形图'
OVERVIEW_DIR = RESULT_DIR / '总览图'
print(RESULT_DIR)
```

## 1. 读取总特征表

```python
summary = pd.read_csv(RESULT_DIR / '全部文件_整体统计和核心特征.csv')
core = pd.read_csv(RESULT_DIR / '用于聚类的核心特征表.csv')
display(summary.head())
display(core.head())
```

## 2. 查看关键统计

```python
display(summary['反应类型初判'].value_counts().rename_axis('反应类型').reset_index(name='数量'))
cols = ['总波峰数量','总波谷数量','ON反应幅度','刺激期相对刺激前变化量','撤光后相对刺激前变化量','波形标准差']
display(summary[cols].describe())
```

## 3. 查看总览图

```python
for name in ['全部预处理波形叠加.png', '预处理波形热图_按刺激响应排序.png', '核心特征分布.png']:
    display(Image(filename=str(OVERVIEW_DIR / name)))
```

## 4. 查看单个细胞的处理结果

```python
example = summary.iloc[0]['file_id']
print('示例:', example)
display(Image(filename=str(PLOT_DIR / f'{example}_预处理峰谷图.png')))
```
